```python
import numpy as np
import torch
from torch import nn, optim
import torch.nn.functional as F
import sys
sys.path.append("../input/")
import d2l_jay9460 as d2l
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')

(corpus_indices, char_to_idx, idx_to_char, vocab_size) = d2l.load_data_jay_lyrics()
```

```python
## GRU简洁实现
# 各参数定义
num_hiddens=256
num_epochs, num_steps, batch_size, lr, clipping_theta = 160, 35, 32, 1e2, 1e-2
pred_period, pred_len, prefixes = 40, 50, ['分开', '不分开']
lr = 1e-2 # 注意调整学习率
# 直接调用pytor.nn模块定义GRU网络
```

```python
gru_layer = nn.GRU(input_size=vocab_size, hidden_size=num_hiddens)
model = d2l.RNNModel(gru_layer, vocab_size).to(device)  # 调用之前的语言模型函数搭建完整的神经网络

d2l.train_and_predict_rnn_pytorch(model, num_hiddens, vocab_size, device,
                                corpus_indices, idx_to_char, char_to_idx,
                                num_epochs, num_steps, lr, clipping_theta,
                                batch_size, pred_period, pred_len, prefixes)
```

```python
## LSTM简洁实现
# 各参数定义（同上，略）
# 直接调用pytor.nn模块定义LSTM网络
lstm_layer = nn.LSTM(input_size=vocab_size, hidden_size=num_hiddens)
model = d2l.RNNModel(gru_layer, vocab_size).to(device)  # 调用之前的语言模型函数搭建完整的神经网络（同上）    
d2l.train_and_predict_rnn_pytorch(model, num_hiddens, vocab_size, device,  # 同上
                                corpus_indices, idx_to_char, char_to_idx,
                                num_epochs, num_steps, lr, clipping_theta,
                                batch_size, pred_period, pred_len, prefixes)
```

```python
# 双向LSTM
lstm_layer = nn.LSTM(input_size=vocab_size, hidden_size=num_hiddens, bidirectional=True)  # bidirectional选择是否双向
model = d2l.RNNModel(lstm_layer, vocab_size).to(device)
```

```python
d2l.train_and_predict_rnn_pytorch(model, num_hiddens, vocab_size, device,
                                corpus_indices, idx_to_char, char_to_idx,
                                num_epochs, num_steps, lr, clipping_theta,
                                batch_size, pred_period, pred_len, prefixes)
```
